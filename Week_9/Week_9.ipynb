# 5. Write a program in python to find alphabet/s having maximum number of instances in a given file.

```python
import string
alpha_dict={s:0 for s in string.ascii_lowercase}
key_with_max_value=list()
with open('/home/syed/Desktop/Mca/Ranga.txt','r+') as f:
    linearray=f.readlines()
    for line in linearray:
        for i in line:
            if i in alpha_dict.keys():
                alpha_dict[i]+=1

max_value=max(alpha_dict.values())
for key,value in alpha_dict.items():
    if value==max_value:
        key_with_max_value.append(key)
print(alpha_dict)
print(f"Key with max value {key_with_max_value} with {max_value} instances.")
```

# 6. A file contains information about programs and courses in the following format: Program,course. Write a Python program to find the number of courses against eachprogram.
### Eg:
### Program,Course
### MCA,Database
### MCA,Java
### M.Sc,Data Structure
### B.Sc, Python
            The output should be: MCA-2, M.Sc-01, B.Sc-01

```python
cource_dict={}
with open('/home/syed/Desktop/Mca/Week_9/courcedata.csv','r+') as f:
    linearray=f.readlines()
    for line in linearray[1:]:
        program,cource=line.split(',')
        if program in cource_dict.keys():
            cource_dict[program]+=1
        else:
            cource_dict[program]=1
for key,value in cource_dict.items():
    print(f"{key} - {value}")
```

# 8. Write a Python program to simulate a Deterministic Finite Automata (DFA) that accepts strings over {0, 1} ending with "01".


```python
def dfa_accepts(input_string):
    state = 0
    for char in input_string:
        if state == 0:
            state = 1 if char == '0' else 0
        elif state == 1:
            state = 2 if char == '1' else 1
        else:
            state = 1 if char == '0' else 0
    return state == 2

# Example usage
print(dfa_accepts("11001"))  # Output: True
print(dfa_accepts("101"))    # Output: True
print(dfa_accepts("100"))    # Output: False
print(dfa_accepts("00000011"))    # Output: False
```
