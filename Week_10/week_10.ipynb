# 5. Write a program in python to find word/s having maximum number of instances in a given file and replace all its occurrences with “Aligarh”.

```python
with open('/home/syed/Desktop/Mca/Ranga.txt', 'r') as file:
    text = file.read()
words = text.split()
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count.setdefault(word, 1)
print(word_count)
max_count=max(word_count.values())
most_frequent_words = [word for word, count in word_count.items() if count == max_count]
print("Word/s with maximum instances:", most_frequent_words, "with", max_count, "occurrences")
for word in most_frequent_words:
    text = text.replace(word, "Aligarh")
with open('/home/syed/Desktop/Mca/Ranga.txt', 'w') as file:
    file.write(text) 
```

# 7. Write a Python program to parse a string to check if it belongs to the language generated by the CFG S → 0S1 | 01 (balanced 0s and 1s).

```python
cfg_rules = {'S': ['0S1', '01']}
def is_in_language(string, symbol='S'):
    if symbol not in cfg_rules:
        return string == ''
    for production in cfg_rules[symbol]:
        if len(production) > len(string):
            continue
        if production == string:
            return True
        if production.startswith('0') and production.endswith('1'):
            inner_string = string[1:-1]
            if is_in_language(inner_string, 'S'):
                return True
    return False
test_strings = ['0011', '000111', '01', '00011111', '10', '0001']
for s in test_strings:
    result = is_in_language(s)
    print(f"String '{s}' belongs to the language: {result}")
```
